## 🔄 Step 1: Model Comparison and Hyperparameter Tuning
In this section, we compare Logistic Regression, Random Forest, SVM, and KNN classifiers using accuracy and classification report.

```
# STEP 1: Compare different models
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

# Splitting data if not already done
X = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\n{name} Results:\n")
    print(classification_report(y_test, y_pred))
```

## 🔍 Step 2: SHAP Explainability
We use SHAP to explain why the model predicts 'Rock' or 'Mine'.

```
# STEP 2: SHAP feature explanation for Random Forest
!pip install shap

import shap
import numpy as np # Import numpy

model = RandomForestClassifier().fit(X_train, y_train)
explainer = shap.TreeExplainer(model)
# Calculate SHAP values for the test set
shap_values = explainer.shap_values(X_test)

# Get the prediction for the first instance in the test set
first_instance = X_test.iloc[[0]]
first_instance_prediction = model.predict(first_instance)[0]

# Determine the index of the predicted class
if first_instance_prediction == 'M':
  predicted_class_index = 1 # 'M' is likely the second class based on previous output
else:
  predicted_class_index = 0 # 'R' is likely the first class

# Convert the input instance to a numpy array
input_instance_numpy = first_instance.values.flatten()


print("Shape of shap_values:", len(shap_values))
if len(shap_values) > 0:
    print("Shape of shap_values[0]:", shap_values[0].shape)
    if shap_values[0].shape[0] > 0:
         print("Shape of shap_values[0][0]:", shap_values[0][0].shape)


print("Shape of input_instance_numpy:", input_instance_numpy.shape)
print("Length of X_test.iloc[0]:", len(X_test.iloc[0]))
print("explainer.expected_value:", explainer.expected_value)

# Check if shap_values has the expected structure (list of 2 arrays with shape (n_instances, n_features))
# If not, assume the structure is a list of n_instances arrays with shape (n_features, n_classes)

if len(shap_values) == 2 and shap_values[0].shape == (X_test.shape[0], X_test.shape[1]):
    # Expected structure for binary classification with TreeExplainer
    shap_values_to_plot = shap_values[predicted_class_index][0, :]
else:
    # Assuming the structure is a list of n_instances arrays with shape (n_features, n_classes)
    # Get SHAP values for the first instance (index 0 in the list) and the predicted class
    if len(shap_values) > 0 and shap_values[0].shape == (X_test.shape[1], len(explainer.expected_value)):
         shap_values_to_plot = shap_values[0][:, predicted_class_index]
    else:
        print("Unexpected shap_values structure.")
        shap_values_to_plot = None # Or handle as an error

if shap_values_to_plot is not None:
    # Visualize the SHAP values for the first prediction of the predicted class
    shap.initjs()
    shap.force_plot(explainer.expected_value[predicted_class_index], shap_values_to_plot, input_instance_numpy)
```
